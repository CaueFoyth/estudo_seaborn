### Plots Matriciais

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
df_flights_matrix = flights.pivot_table(index="year", columns="month", values="passengers", observed=True)
df_flights_matrix
## Organizando melhor a tabela passando o index o ano, com suas colunas os meses e com os valores o nÃºmero de passageiros
```

```python
fig, ax = plt.subplots(figsize=(15,5)) ## Mudar o tamanho do grÃ¡fico
sns.heatmap(df_flights_matrix, ax=ax, annot=True, fmt=".0f", linewidths=1, cmap="magma") 
## ax = tamanho, annot = colocar valor dentro da celula, fmt = formataÃ§Ã£o, linewidths = linha entre as celulas, cmap = tema
plt.show()
```

```python
sns.clustermap(df_flights_matrix)
plt.show()
```
