## Estudo sobre Visualização de dados avançada

### plots de distribuição e categórica

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(tips['total_bill'], kde=True) #Valores concentrados
plt.show()
```

```python
sns.displot(tips['total_bill'], kde=False, bins=30) #Valores concentrados, tirando linhas e aumentando a dispersao em colunas
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg") #Gráfico de dispersão entre total de gorjeta com a coluna do valor total da conta (kind == tipo de grafico)
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") #Gráfico de dispersão entre total de gorjeta com a coluna do valor total da conta (kind == tipo de grafico)
plt.show()
```

```python
sns.pairplot(tips, hue="sex", palette="coolwarm")
plt.show()
#plotando uam correlação das variaveis numericas com variaveis categoricas(sexo)
#hue == para variaveis categoricas
#palette == tema dos graficos
```

```python
sns.rugplot(tips["total_bill"])
plt.show()
# Forma para ver concentração de dados
```

```python
sns.kdeplot(tips["total_bill"])
sns.rugplot(tips["total_bill"])
plt.show()
# Forma para ver concentração de dados de uma maneira diferente, linha é a estivmativa das aparições abaixo
```

```python
sns.barplot(x= "sex", y="total_bill", data=tips)
plt.show()
# Media
```

```python
import numpy as np
```

```python
sns.barplot(x= "sex", y="total_bill", data=tips, estimator=np.std)
plt.show()
# Mostrar disvio padrão
```

```python
sns.countplot(x= "sex", data=tips)
plt.show()
# Contagem dos sexos
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="day", palette="rainbow")
plt.show()
# Grafico de distribuição, calculo por percentil de dados, bolinas fora dos valores padrão, linha do meio dos quadrados é a mediana, linhas dos cantos maxima e minima
```

```python
sns.boxplot(x="total_bill", y="day", data=tips, hue="day", palette="rainbow", orient="h")
plt.show()
# Grafico de distribuição, calculo por percentil de dados, bolinas fora dos valores padrão, linha do meio dos quadrados é a mediana, linhas dos cantos maxima e minima
```

```python
sns.boxplot(x="total_bill", y="day", data=tips, hue="sex", palette="rainbow", orient="h")
plt.show()
# Grafico de distribuição, calculo por percentil de dados, bolinas fora dos valores padrão, linha do meio dos quadrados é a mediana, linhas dos cantos maxima e minima
```

```python
sns.violinplot(x="total_bill", y="day", data=tips, hue="sex", palette="rainbow", split=True)
plt.show()
# mesma ideia dos dados do boxplot porem em uma visualização de violino dividido por sexo
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.show()
```

### Plots de regressão

```python
tips
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", markers=['o', 'v'], scatter_kws={'s':100})
plt.show()
## Markers = mudar os pontos
## scatter_kws={'s':100} = Aumentar o tamnho dos marcadores
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", col="time", row="day")
plt.show()
## Quebrando os dados em duas colunas por tempo(almoço e janta) e em linha por dia da semana
```
